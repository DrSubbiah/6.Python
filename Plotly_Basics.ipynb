```python
import plotly.express as px
import pandas as pd
```

```python
birth_wt_da = pd.read_csv("/content/BirthWt.csv") #/content/BirthWt.csv
```

```python
birth_wt_da.head(5)
```

```python
birth_wt_da.dtypes
```

```python
birth_wt_da.shape
```

```python
birth_wt_da.columns
```

```python
# Convert  attributes to object
cols = ['id', 'matagegp', 'gestcat']
birth_wt_da[cols] = birth_wt_da[cols].astype('object')
```

```python
px.scatter(birth_wt_da,x='gestwks',y='bweight')
#plt.scatter("Weeks")
#plt.ylabel("Birth Weight")
```

```python
px.scatter(birth_wt_da,x='gestwks',y='bweight',color="sex",
          labels=dict(gestwks="Gestation Week", bweight="BirthWeight", sex="Baby Gender"))
```

```python
px.histogram(birth_wt_da,x='bweight')
```

```python
px.histogram(birth_wt_da,x='bweight',color="matagegp")
```

```python
px.box(birth_wt_da,x='bweight')
```

```python
px.box(birth_wt_da,y='bweight')
```

```python
px.box(birth_wt_da,y='bweight',color="matagegp")
```

```python
px.histogram(birth_wt_da,x='sex',color="matagegp")
```

```python
px.histogram(birth_wt_da,x='sex',color = 'sex',color_discrete_map = {'female':'red','male':'blue'})
```

```python
px.histogram(birth_wt_da,x='sex',color="matagegp")
```

```python
px.bar(birth_wt_da,x='matagegp')
```

```python
birth_wt_da['matagegp'] = birth_wt_da['matagegp'].astype('object')
```

```python
px.bar(birth_wt_da,x='matagegp',barmode='group')
```

```python
px.bar(birth_wt_da,x='matagegp',barmode='group',color='sex')
```
