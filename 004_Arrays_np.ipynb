#### This is a 1-D array

```python
arr1d=np.array([99,102,199])
print(arr1d)
print(arr1d.ndim)
print(arr1d.shape)
```

#### Following two are examples of  2-D array

```python
arr2d = np.array([[89,10], [89, 13], [182,17]])
print(arr2d)
print(arr2d.ndim)
print(arr2d.shape)
print(arr2d.dtype)
```

```python
arr2d1 = np.array([[89,89,182], [10, 13,17]])
print(arr2d1)
print(arr2d1.ndim)
print(arr2d1.shape)
```

```python
arr3d = np.array([[[44,45], [5,5]], [[43,46], [4,9]],[[90,92],[7,10]]])

print(arr3d.ndim)
print(arr3d.shape)
print(arr3d)
```
